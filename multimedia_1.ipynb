# Adding random noise to the dataset

```python
import random
print("Original NumPy Array:")
print(X_new)

# Create a new array for noisy data
X_new_noisy = np.copy(X_new)

# Add random noise to each element
for i in range(X_new.shape[0]):
    for j in range(X_new.shape[1]):
        X_new_noisy[i, j] += random.uniform(0.1, 0.12)

# Display the NumPy array with added random noise
print("\nNumPy Array with Added Noise:")
print(X_new_noisy)
```

# Scaling the datasets with StandardScaler 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scale=scaler.fit_transform(X_new_noisy)
X_scale
```

# Implementing Kfold

```python
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
kf
```

```python
def get_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
get_score(SVC(),X_train,X_test,y_train,y_test)
```

```python
from sklearn.model_selection import StratifiedKFold#stratified and kfold is similar but stratified uniformly distribute the datasets
fold=StratifiedKFold(n_splits=5)
fold
```

```python
score_lr=[]
score_svm=[]
score_lrf=[]
for train_index,test_index in kf.split(X_scale):
    X_train,X_test,y_train,y_test=X_scale[train_index],X_scale[test_index],y_new[train_index],y_new[test_index]
    score_lr.append(get_score(LogisticRegression(),X_train,X_test,y_train,y_test))
    score_svm.append(get_score(SVC(),X_train,X_test,y_train,y_test))
    score_lrf.append(get_score(RandomForestClassifier(n_estimators=40),X_train,X_test,y_train,y_test))
        
```
